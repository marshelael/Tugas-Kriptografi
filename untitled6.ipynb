```python
def prepare_plaintext(name):
    name = name.upper().replace(" ", "")
    keyword = "KRIPTOGRAFI"
    if len(name) < 15:
        name += keyword[:15 - len(name)]
    return name[:15]
```

```python
def substitusi_encrypt(text, key=5):
    hasil = ""
    for ch in text:
        if ch.isalpha():
            hasil += chr((ord(ch) - 65 + key) % 26 + 65)
        else:
            hasil += ch
    return hasil

def substitusi_decrypt(text, key=5):
    hasil = ""
    for ch in text:
        if ch.isalpha():
            hasil += chr((ord(ch) - 65 - key) % 26 + 65)
        else:
            hasil += ch
    return hasil
```

```python

def ekspansi_encrypt(text):
    vowels = "AEIOU"
    if text[0] in vowels:
        return "INI" + text
    else:
        return text[1:] + text[0] + "ITU"

def ekspansi_decrypt(text):
    if text.startswith("INI"):
        return text[3:]
    elif text.endswith("ITU"):
        t = text[:-3]
        return t[-1] + t[:-1]
    else:
        return text
```

```python
def encrypt_process(name):
    print("\n=== PROSES ENKRIPSI ===")
    plaintext = prepare_plaintext(name)
    print("Plaintext (15 karakter):", plaintext)

    sub = substitusi_encrypt(plaintext)
    print("1ï¸âƒ£ Hasil Substitusi:", sub)

    blocks = blocking(sub)
    perm_blocks = [permutasi_encrypt(b) for b in blocks]
    perm_result = ''.join(perm_blocks)
    print("2ï¸âƒ£ Hasil Permutasi:", perm_result)

    expanded = ekspansi_encrypt(perm_result)
    print("3ï¸âƒ£ Hasil Ekspansi (Ciphertext):", expanded)
    return expanded
```

```python
def decrypt_process(cipher):
    print("\n=== PROSES DEKRIPSI ===")
    rev_exp = ekspansi_decrypt(cipher)
    print("1ï¸âƒ£ Balikkan ekspansi:", rev_exp)

    blocks = blocking(rev_exp)
    perm_blocks = [permutasi_decrypt(b) for b in blocks]
    perm_rev = ''.join(perm_blocks)
    print("2ï¸âƒ£ Balikkan permutasi:", perm_rev)

    sub_rev = substitusi_decrypt(perm_rev)
    print("3ï¸âƒ£ Balikkan substitusi:", sub_rev)
    print("\nðŸ“œ Plaintext hasil dekripsi:", sub_rev)
    return sub_rev
```

```python
name = "MARSHELA ELSY PUTRI"
cipher = encrypt_process(name)
plain = decrypt_process(cipher)
```

```python
def blocking(text, block_size=3):
    return [text[i:i+block_size] for i in range(0, len(text), block_size)]

def permutasi_encrypt(block):
    # Simple permutation: swap the first and last characters
    if len(block) == 3:
        return block[2] + block[1] + block[0]
    return block # Return as is if block size is not 3

def permutasi_decrypt(block):
    # Simple permutation: swap the first and last characters
    if len(block) == 3:
        return block[2] + block[1] + block[0]
    return block # Return as is if block size is not 3
```
